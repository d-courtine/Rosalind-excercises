# Permutations

```python
import numpy as np
```

```python
def permutation(list1):
    # If the length of list = 0 no permuatations possible
    if len(list1) == 0:
        return []
    # If the length of list=1, return that element
    if len(list1) == 1:
        return [list1]
    l = []
    
    for i in range(len(list1)):
        # Extract list1[i] ==> Will become the first element of the main / any sublists
        m = list1[i]
        
        # The remaining list
        remlist1 = list1[:i] + list1[i+1:]

        # Generating all permutations where m is first element
        # ==> For the main list and all the sublists possible
        for p in permutation(remlist1):
            l.append([m] + p)  
    return l
```

```python
k = 5 #Value given by Rosalind

mylist = list(np.linspace(1, k, k, dtype=np.int64))

result = permutation(mylist)
```

```python
print(len(result))
for i in result:
    for j in i:
        print(j, end = ' ')
    print()
```
