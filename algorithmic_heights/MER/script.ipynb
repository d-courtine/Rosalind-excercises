Initialize the two list to merge, _a.k.a._ read them from the file:

```python
a = list()
b = list()

num_line = 1
# with open('test.txt', 'r') as f:
with open('rosalind_mer.txt', 'r') as f:

    lines = f.readlines()
    for line in lines:
        if num_line == 2:
            a = line.rstrip().split(' ')
        elif num_line == 4:
            b = line.rstrip().split(' ')
        num_line += 1
```

Merge the two ordered list:

```python
merge = list()
final_len = len(a) + len(b)
```

The main loop

```python
iteration = 0
while len(merge) != final_len :
    #append the last elements to the "merge" list
    if len(a) == 0:
        merge = merge + b
    elif len(b) == 0:
        merge = merge + a
        
    #compare the first element of a or b to populate the "merge" list
    else:
        if int(a[0]) > int(b[0]):
            merge.append(b.pop(0))
        else:
            merge.append(a.pop(0))
    iteration += 1
```

```python
for i in merge:
    print(i, end = ' ')
```
